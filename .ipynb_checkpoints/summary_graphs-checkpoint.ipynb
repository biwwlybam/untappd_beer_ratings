```python
# import necessary modules
import pandas as pd
from datetime import date
import matplotlib.pyplot as plt
import numpy as np
```

```python
# load most up-to-date scores (assuming updated today)
today_str = date.today().strftime('%m-%d-%Y')
df = pd.read_csv('burial_untappd_'+today_str+'.csv', index_col=[0])
```

### Function to plot inexact names

```python
def single_beer_plot(df, beer_name):
    # find scores for specified beer and transpose so they're in a column, skipping name and style
    # also set case arg to False so not case sensitive 
    scores = (df[df['beer_names']
        .str.contains(beer_name, case=False)]
        .T[2:]
    )
    
    # get beer names for legend
    # transpose to cut to what we need and transpose back
    names = (df[df['beer_names']
            .str.contains(beer_name, case=False)]
            .T[:1]
            .T['beer_names']
            )
    
    # plot scores by date
    plt.plot(scores.index, scores)
    # use title case of beer name arg as title
    plt.title(beer_name.title())
    plt.legend(names)
    plt.xlabel('Date')
    plt.ylabel('Score')
    plt.xticks(rotation = 45)
    plt.show()
    plt.clf()
```

### Function to plot exact name

```python
def single_beer_plot_exact(df, beer_name):
    # find scores for specified beer and transpose so they're in a column, skipping name and style
    # also set case arg to False so not case sensitive 
    scores = df[df['beer_names'] == beer_name].T[2:]
    
    # plot scores by date
    plt.plot(scores.index, scores)
    plt.title(beer_name)
    plt.xlabel('Date')
    plt.ylabel('Score')
    plt.xticks(rotation = 45)
    plt.show()
    plt.clf()
```

### Examples

```python
single_beer_plot(df, 'bright')
```

```python
single_beer_plot_exact(df, 'Surf Wax IPA')
```
